## Data Cleaning

# Problems noticed :
  -year has so many non year values

  -some price has non-numeric values(Ask For Price)

  -covert price and year to int by removing comma (object to int)

  -kms_driven has object values with kms at last.

  -It has nan values and some non km values

  -fuel_type has nan values

  -Keep first three words of the name

```python
# year has many non-year values
df = df[df['year'].str.isnumeric()]
```

```python
# year is in object --> Change to integer
df['year'] = df['year'].astype(int)
```

```python
# Price has Ask for Price
df = df[df['Price']!='Ask For Price']
```

```python
# Price has commas in its prices and is in object
df['Price'] = df['Price'].str.replace(',','').astype(int)
```

```python
# kms_driven has object values with kms at last.
df['kms_driven'] = df['kms_driven'].str.split().str.get(0).str.replace(',','')
```

```python
# It has nan values and some non km values
df = df[df['kms_driven'].str.isnumeric()]
```

```python
# Change to integer
df['kms_driven'] = df['kms_driven'].astype(int)
```

```python
df.isnull().sum()
```

```python
#fuel_type has nan values
df=df[~df['fuel_type'].isna()]
```

```python
df.info()
```

```python
# Keep first three words of the name
df['name'] = df['name'].str.split().str.slice(start=0,stop=3).str.join(' ')
```

```python
# Resetting the index of the final cleaned data
df = df.reset_index(drop=True)
```

## Data Preprocessing

```python
x = df.drop(['Price'], axis=1)
y = df['Price']
```

```python
x.head()
```

```python
y.head()
```

```python
#One-hot-encoding
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import make_column_transformer

ohe = OneHotEncoder()
ohe.fit(x[['name','company','fuel_type']])

ct = make_column_transformer((OneHotEncoder(categories = ohe.categories_),['name','company','fuel_type']),
                                    remainder='passthrough')
```

## Train-Test Split

```python
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test = train_test_split(x ,y, test_size= 0.2, random_state = 0)

print("Training set size :", len(x_train))
print("Test set size :", len(x_test))
```

## Model Training and Evaluation

```python
from sklearn.pipeline import make_pipeline
from sklearn.metrics import r2_score
```

```python
# Linear Regression Model
from sklearn.linear_model import LinearRegression
lr = LinearRegression()

pipe = make_pipeline(ct, lr)

pipe.fit(x_train, y_train)
```

```python
y_pred = pipe.predict(x_test)
r2 = r2_score(y_test,y_pred)
r2 = round(r2, 2)
print("Linear Regression")
print("R^2 score :", r2)
```

```python
# Random Forest Regression Model
from sklearn.ensemble import RandomForestRegressor
rf = RandomForestRegressor(n_estimators= 100, criterion='squared_error',n_jobs =-1, random_state = 0)

pipe = make_pipeline(ct,rf)

pipe.fit(x_train,y_train)
```

```python
y_pred = pipe.predict(x_test)
r2 = r2_score(y_test,y_pred)
r2 = round(r2, 2)
print("Random Forest")
print("R^2 score :", r2)
```

```python
# XGBoost Regression Model
import xgboost as xgb
xgbr = xgb.XGBRegressor(n_estimators = 100, objective = "reg:squarederror",n_jobs =-1, random_state = 0)

pipe = make_pipeline(ct,xgbr)

pipe.fit(x_train,y_train)
```

```python
y_pred = pipe.predict(x_test)
r2 = r2_score(y_test,y_pred)
r2 = round(r2, 2)
print("XGBoost")
print("R^2 score :", r2)
```

# Hyperparameter Tuning

```python
from sklearn.model_selection import GridSearchCV

# Define the hyperparameter grid
param_grid = {
    'xgbregressor__learning_rate': [0.1, 0.2, 0.3],
    'xgbregressor__n_estimators': list(range(500, 1001, 50)),
    'xgbregressor__objective': ["reg:squarederror"],
    'xgbregressor__max_depth': [2, 3, 4],
    'xgbregressor__reg_lambda': [1.0, 1.5, 2.0],
    'xgbregressor__n_jobs': [-1],
    'xgbregressor__random_state': [0]
}
```

```python
# Create the GridSearchCV object
grid_search = GridSearchCV(pipe, param_grid, cv=5)
```

```python
x_train
```

```python
# Fit the model using GridSearchCV
grid_search.fit(x_train, y_train)
```

```python
# Get the best parameters and best score
best_params = grid_search.best_params_
```

```python
# Use the best model for predictions
best_model = grid_search.best_estimator_
predictions = best_model.predict(x_test)
```

```python
print(best_params)
```

```python
best_r2 = r2_score(y_test,predictions)
best_r2 = round(best_r2, 2)
print("XGBoost")
print("Best R^2 score :",best_r2)
```
